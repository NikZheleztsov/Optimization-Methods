### Постановка задачи

На интервале $[x_{min}, x_{max}]$ задан сигнал $f_k = f(x_k)$, где дискретная последовательность отсчетов $x_k = x_{min} + k(x_{max} - x_{min})/K$, где $k = 0, ..., K,$ а $K$ - количество отсчетов. На сигнал наложен дискретный равномерный шум $\sigma = (\sigma_0, ..., \sigma_K)$ с нулевым средним и амлитудой, равномерно распределенной на интервале $[-a, a]:$ $\tilde f_k = f_k + \sigma_k,$ $\sigma_k = rnd(-a, a)$. 

Необходимо осуществить фильтрацию сигнала с использованием "Манхеттенской" метрики методом квадратичного взвешенного скользящего среднего

### Ход выполнения работы

#### Исходные данные
$$ f_k = sin(x_k) + 0,5, $$
$$ x_k = x_{min} + k(x_{max} - x_{min}) / K $$
$$ x_{min} = 0, x_{max} = \pi $$
$$ k = 0, ..., K,$$
$$  K = 100 $$

Амплитуда равномерного шума: $2a = 0,5 $
<br> Дискретизация веса свертки: $ \lambda_l = l / L, l = 0, ..., L, L = 10 $
<br> Вероятность попадания в окрестность экстремума: $P = 0,95 $
<br> Интервал неопределенности: $\epsilon = 0,01$
<br> Размер скользящего окна: $r = 3, r = 5$

#### Построение графиков

Построим график заданной функции f(x):

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
K = 100
x = np.array([k * np.pi / K for k in range(K + 1)])
y = np.sin(x) + 0.5

plt.plot(x, y, label='f(x) = sin(x) + 0.5')
plt.legend(loc = 'lower center')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
```

Добавим шум в функцию и построим график зашумленной функции. Согласно условию задачи требуется, чтобы амплитуда шума была равномерно распределена

```python
# Generating of uniform destributed sigma
np.random.seed(123)
sigma = np.array([np.random.uniform(-0.25, 0.25, K + 1)]).reshape(101, )

# Noisy f(x)
y_noisy = y + sigma

# Plot
plt.plot(x, y, label='f(x) = sin(x) + 0.5')
plt.plot(x, y_noisy, label='noise')
plt.legend(loc = 'lower center')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
```

По условию задания необходимо использовать среднее квадратичное для формулы взвешенного скользящего среднего:
$$ \bar f_k(\alpha) = \sqrt{\sum_{j = k - M}^{k + M} \tilde f_j^2 \alpha_{j + M + 1 - k}} $$

```python
# Взвешенное скользящее среднее
def filtered_signal(y, r, alpha):
    M = (r - 1) // 2
    y_pad = np.pad(y, (M, M), 'constant', constant_values = 1)
    filter_y = np.array([np.sqrt([np.sum([np.power(y_pad[i + j], 2) * a \
                                for j, a in enumerate(alpha)])
                                for i, _ in enumerate(y)])]).reshape(-1, 1)
    return filter_y
```

В качестве критерия зашумленности ($\omega$) и критерия отличия ($\delta$) используется "Манхеттенская метрика":
$$ \omega = \sum_{k = 1}^{K} | \bar f_k - \bar f_{k - 1} | $$
$$ \delta = \dfrac {1}{K} \sum_{k = 1}^{K} | \bar f_k - \tilde f_{k - 1} | $$

```python
# Omega
def noisiness(filtered_y):
    K = filtered_y.shape[0]
    return np.sum(np.abs([filtered_y[k] - filtered_y[k - 1] \
                   for k in range (1, K)]))

# Delta
def dif (filtered_y, noisy_y):
    K = filtered_y.shape[0]
    return (1 / (K - 1)) * np.sum(np.abs([filtered_y[k] - noisy_y[k] \
                           for k in range (K)]))
```

Для нахождения расстояния орт приближенно найденного оптимального значения до идеальной точки используется также "Манхеттенская" метрика:

$$ dist = |\omega| + |\delta| $$

```python
def dist(omega, delta):
    return np.absolute(omega) + np.absolute(delta)
```

Случайные значения весов берутся симметричными относительно центрального веса $a_{M + 1}$ и рассчитываются с учетом условия нормировки,т.е. равномерно распределены на последовательности остаточных
интервалов:

$$ a_{M + 1} = rnd(0, 1), $$
$$ a_m = a_{r - m + 1} = 0.5 rnd(0.1 - \sum_{s = m + 1}^{r - m} a_s), m = 2, ..., M $$
$$ a_1 = a_r = 0.5 (1 - \sum_{s = 2}^{r - 1} a_s) $$

```python
def generate_random_alpha(r):
    M = (r - 1) // 2
    alpha = np.zeros(r)
    alpha[M] = np.random.random()
    
    if r > 3:
        for m in range(2, M + 1):
            alpha[m - 1] = alpha[r - m] = \
                0.5 * np.random.uniform(0.0, 1 - np.sum( [alpha[s] \
                                        for s in range(m, r - m - 1)]))

    alpha[0] = alpha[r - 1] = \
        0.5 * (1 - np.sum([alpha[s] for s in range(1, r - 1)]))
    
    return alpha
```

Необходимо применить линейную свертку критериев:
$$ J = \lambda\omega + (1 - \lambda)\delta \to min $$
Значения весов $\lambda_l = \dfrac{l}{L}, l = 0, ..., 10 $

Число испытаний оценивается по формуле:
$$ N = \dfrac{ln(1 - P)}{ln(1 - \dfrac{\epsilon}{x_{max} - x_{min}})} $$

Вычислим число испытаний:

```python
P = 0.95
eps = 0.01
x_min = 0
x_max = np.pi

N = int(np.log(1 - P) // np.log(1 - eps / (x_max - x_min)))
print("Number of tests: ", N)
```

Напишем функцию для метода случайного поиска:

```python
def random_search_method(noisy_y, r, lam, debug = False, N = N):
    best_alpha = np.array([])
    best_dist = np.inf
    best_omega = np.inf
    best_delta = np.inf
    best_J = np.inf
    best_f = []
    
    if (debug):
        plt.plot(x, y, label='f(x) = sin(x) + 0.5')
        plt.plot(x, y_noisy, label='noise')
        plt.xlabel('x')
        plt.ylabel('f(x)')
    
    for i in range(N):
        alpha = generate_random_alpha(r)
        temp_f = filtered_signal(noisy_y, r, alpha)
        temp_omega = noisiness(temp_f)
        temp_delta = dif(temp_f, noisy_y)
        temp_dist = dist(temp_omega, temp_delta)
        J = lam * temp_omega + (1 - lam) * temp_delta
        
        if (J < best_J):
            best_alpha = alpha
            best_dist = temp_dist
            best_omega = temp_omega
            best_delta = temp_delta
            best_J = J
            best_f = temp_f
            
    if (debug):
        temp_val = int(np.ceil(r / 3.0))
        plt.plot(x[temp_val : 101 - temp_val], best_f, label = 'filtered')
        plt.legend(loc = 'lower center')
        plt.show()

    return [best_dist, best_alpha, best_omega, best_delta, best_J]
```

```python
import pandas as pd

def experiment(r, print_data = False):
    data = []
    lambd = np.linspace(0.0, 1.0, 11)
    for lam in lambd:
        temp_data = random_search_method(y_noisy, r, lam)
        temp_data[1] = temp_data[1].round(4)
        temp_data.insert(0, lam)
        data.append(temp_data)

    columns = ['lambda', 'dist', 'alpha', 'noisiness', 'difference', 'J']
    frame = pd.DataFrame(data, columns=columns)
    if (print_data):
        display(frame)
        print('Most optimal lambda:')
        display(frame.loc[[frame['dist'].idxmin()], :].drop('alpha',\
                                                            axis = 1))

    alpha = frame.loc[[frame['dist'].idxmin()], :]['alpha'].values[0]
    filtered_y = filtered_signal(y_noisy, r, alpha)
    
    fig, ax = plt.subplots(2, 1, figsize=(10,10))
    
    temp_value = (101 - filtered_y.shape[0]) // 2
    ax[0].plot(x, y, label='f(x) = sin(x) + 0.5')
    ax[0].plot(x, y_noisy, label='noise')
    # ax[0].plot(x[temp_value: 101 - temp_value], filtered_y, label='filtered')
    ax[0].plot(x, filtered_y, label='filtered')
    ax[0].legend(loc = 'lower center')
    ax[0].set_xlabel('x')
    ax[0].set_ylabel('f(x)')
    ax[0].grid()
    
    colors = ['bo', 'go', 'ro', 'co', 'mo', 'yo', 'ko']
    ax[1].grid()
    ax[1].set_xlabel('omega')
    ax[1].set_ylabel('delta')
    for i in range(lambd.size):
        ax[1].plot(frame.loc[i, 'noisiness'], \
                   frame.loc[i, 'difference'], colors[i % 7])
        
    plt.show()
```

#### Проведем эксперимент для скользящего окна r = 3:

```python
experiment(3, True)
```

#### Проведем эксперимент для скользящего окна r = 5:

```python
experiment(5, True)
```

### Проведем эксперименты для различных скользящих окон:

```python
for r in range(1, 20):
    print('r =', r)
    experiment(r)
```
